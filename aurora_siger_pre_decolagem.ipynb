## 1.1 – Organização e Descrição da Telemetria

A telemetria da nave Aurora monitora continuamente os seguintes parâmetros críticos durante a fase de pré-decolagem:

| Parâmetro | Descrição | Unidade | Faixa Segura |
|---|---|---|---|
| Temperatura Interna | Temperatura interna da cabine e dos compartimentos | °C | 15°C a 28°C |
| Temperatura Externa | Temperatura do ambiente externo à nave | °C | -60°C a -10°C |
| Integridade Estrutural | Verificação de integridade do casco e estrutura | Binário (0/1) | 1 (OK) |
| Nível de Energia | Percentual de carga dos módulos de bateria | % | ≥ 70% |
| Pressão dos Tanques | Pressão interna dos tanques de propelente | kPa | 300 a 400 kPa |
| Módulo de Propulsão | Status operacional do sistema de propulsão | Binário (0/1) | 1 (Operacional) |
| Módulo de Navegação | Status operacional do sistema de navegação | Binário (0/1) | 1 (Operacional) |
| Módulo de Comunicação | Status operacional do sistema de comunicação | Binário (0/1) | 1 (Operacional) |

````python
# ============================================================
# MISSÃO AURORA SIGER – Sistema de Verificação Pré-Decolagem
# FIAP – Fase 1: Decolagem da Missão
# Aluno: Paulo Roberto Faulstich Rego | Grupo 58
# ============================================================

# ============================================================
# 1.1 – Leitura dos Dados de Telemetria
# ============================================================

print("=" * 65)
print("   NAVE AURORA – LEITURA DOS DADOS DE TELEMETRIA")
print("=" * 65)

# Dados coletados pelos sensores embarcados da nave Aurora
temperatura_interna   = 22.5    # graus Celsius – temperatura da cabine
temperatura_externa   = -45.2   # graus Celsius – temperatura orbital externa
integridade_estrutural = 1      # 1 = integra, 0 = falha detectada
nivel_energia         = 87.3    # % de carga atual dos bancos de bateria
pressao_tanques       = 342.0   # kPa – pressão dos tanques de propelente
modulo_propulsao      = 1       # 1 = operacional, 0 = falha
modulo_navegacao      = 1       # 1 = operacional, 0 = falha
modulo_comunicacao    = 1       # 1 = operacional, 0 = falha

print(f"  Temperatura Interna      : {temperatura_interna} °C")
print(f"  Temperatura Externa      : {temperatura_externa} °C")
print(f"  Integridade Estrutural   : {integridade_estrutural} (1=OK, 0=Falha)")
print(f"  Nivel de Energia         : {nivel_energia} %")
print(f"  Pressao dos Tanques      : {pressao_tanques} kPa")
print(f"  Modulo de Propulsao      : {modulo_propulsao} (1=Operacional, 0=Falha)")
print(f"  Modulo de Navegacao      : {modulo_navegacao} (1=Operacional, 0=Falha)")
print(f"  Modulo de Comunicacao    : {modulo_comunicacao} (1=Operacional, 0=Falha)")
print("=" * 65)
````

## 1.2 – Algoritmo de Verificação

O algoritmo de verificação avalia cada parâmetro de telemetria contra as faixas seguras predefinidas e decide entre **PRONTO PARA DECOLAR** ou **DECOLAGEM ABORTADA**.

### Pseudocódigo

```
ALGORITMO VerificacaoPreDecolagem

  VARIAVEIS
    temperatura_interna, temperatura_externa: real
    integridade_estrutural: inteiro
    nivel_energia: real
    pressao_tanques: real
    modulo_propulsao, modulo_navegacao, modulo_comunicacao: inteiro
    falhas: lista de texto

  INICIO
    // Leitura dos dados
    LER temperatura_interna
    LER temperatura_externa
    LER integridade_estrutural
    LER nivel_energia
    LER pressao_tanques
    LER modulo_propulsao
    LER modulo_navegacao
    LER modulo_comunicacao

    falhas <- lista_vazia

    // Verificacoes
    SE temperatura_interna < 15 OU temperatura_interna > 28 ENTAO
      ADICIONAR "Temperatura Interna" a falhas
    FIM_SE

    SE temperatura_externa < -60 OU temperatura_externa > -10 ENTAO
      ADICIONAR "Temperatura Externa" a falhas
    FIM_SE

    SE integridade_estrutural <> 1 ENTAO
      ADICIONAR "Integridade Estrutural" a falhas
    FIM_SE

    SE nivel_energia < 70 ENTAO
      ADICIONAR "Nivel de Energia" a falhas
    FIM_SE

    SE pressao_tanques < 300 OU pressao_tanques > 400 ENTAO
      ADICIONAR "Pressao dos Tanques" a falhas
    FIM_SE

    SE modulo_propulsao <> 1 ENTAO
      ADICIONAR "Modulo de Propulsao" a falhas
    FIM_SE

    SE modulo_navegacao <> 1 ENTAO
      ADICIONAR "Modulo de Navegacao" a falhas
    FIM_SE

    SE modulo_comunicacao <> 1 ENTAO
      ADICIONAR "Modulo de Comunicacao" a falhas
    FIM_SE

    // Decisao final
    SE TAMANHO(falhas) == 0 ENTAO
      IMPRIMIR "PRONTO PARA DECOLAR"
    SENAO
      IMPRIMIR "DECOLAGEM ABORTADA"
      PARA CADA falha EM falhas FACA
        IMPRIMIR "Falha: " + falha
      FIM_PARA
    FIM_SE

  FIM
```

### Fluxograma

```
         [INICIO]
             |
      [Ler dados de telemetria]
             |
             v
   <Temp. Interna entre 15 e 28?> ---NAO---> [Registrar falha: Temp. Interna]
             | SIM                                         |
             v                                             |
   <Temp. Externa entre -60 e -10?> --NAO--> [Registrar falha: Temp. Externa]
             | SIM                                         |
             v                                             |
   <Integridade Estrutural == 1?> --NAO---> [Registrar falha: Integridade]
             | SIM                                         |
             v                                             |
   <Nivel de Energia >= 70%?> ------NAO---> [Registrar falha: Energia]
             | SIM                                         |
             v                                             |
   <Pressao entre 300 e 400 kPa?> --NAO---> [Registrar falha: Pressao]
             | SIM                                         |
             v                                             |
   <Todos os modulos == 1?> --------NAO---> [Registrar falha: Modulo(s)]
             | SIM                                         |
             v                                             |
         <Ha falhas?> <----------------------------------<-+
          /       \
        NAO       SIM
         |         |
         v         v
    [PRONTO    [DECOLAGEM
      PARA      ABORTADA]
    DECOLAR]   [Listar falhas]
         |         |
         +----+----+
              |
            [FIM]
```

## 1.3 – Script Python: Execução das Verificações

````python
# ============================================================
# 1.2 / 1.3 – Faixas Seguras e Algoritmo de Verificação
# ============================================================

# Limites operacionais seguros definidos pela equipe de missão
TEMP_INT_MIN   = 15.0    # °C
TEMP_INT_MAX   = 28.0    # °C
TEMP_EXT_MIN   = -60.0   # °C
TEMP_EXT_MAX   = -10.0   # °C
ENERGIA_MIN    = 70.0    # %
PRESSAO_MIN    = 300.0   # kPa
PRESSAO_MAX    = 400.0   # kPa

print("=" * 65)
print("   NAVE AURORA – VERIFICACAO DOS SISTEMAS")
print("=" * 65)

falhas = []  # Lista para registrar sistemas fora do limite

# --- Verificacao: Temperatura Interna ---
if temperatura_interna >= TEMP_INT_MIN and temperatura_interna <= TEMP_INT_MAX:
    print(f"  [OK] Temperatura Interna     : {temperatura_interna} C  (faixa: {TEMP_INT_MIN} a {TEMP_INT_MAX} C)")
else:
    print(f"  [FALHA] Temperatura Interna  : {temperatura_interna} C  FORA DO LIMITE!")
    falhas.append("Temperatura Interna")

# --- Verificacao: Temperatura Externa ---
if temperatura_externa >= TEMP_EXT_MIN and temperatura_externa <= TEMP_EXT_MAX:
    print(f"  [OK] Temperatura Externa     : {temperatura_externa} C  (faixa: {TEMP_EXT_MIN} a {TEMP_EXT_MAX} C)")
else:
    print(f"  [FALHA] Temperatura Externa  : {temperatura_externa} C  FORA DO LIMITE!")
    falhas.append("Temperatura Externa")

# --- Verificacao: Integridade Estrutural ---
if integridade_estrutural == 1:
    print(f"  [OK] Integridade Estrutural  : {integridade_estrutural}  (estrutura integra)")
else:
    print(f"  [FALHA] Integridade Estrutural: {integridade_estrutural}  FALHA DETECTADA!")
    falhas.append("Integridade Estrutural")

# --- Verificacao: Nivel de Energia ---
if nivel_energia >= ENERGIA_MIN:
    print(f"  [OK] Nivel de Energia        : {nivel_energia}%  (minimo: {ENERGIA_MIN}%)")
else:
    print(f"  [FALHA] Nivel de Energia     : {nivel_energia}%  INSUFICIENTE!")
    falhas.append("Nivel de Energia")

# --- Verificacao: Pressao dos Tanques ---
if pressao_tanques >= PRESSAO_MIN and pressao_tanques <= PRESSAO_MAX:
    print(f"  [OK] Pressao dos Tanques     : {pressao_tanques} kPa  (faixa: {PRESSAO_MIN} a {PRESSAO_MAX} kPa)")
else:
    print(f"  [FALHA] Pressao dos Tanques  : {pressao_tanques} kPa  FORA DO LIMITE!")
    falhas.append("Pressao dos Tanques")

# --- Verificacao: Modulos Criticos ---
modulos = {
    "Modulo de Propulsao" : modulo_propulsao,
    "Modulo de Navegacao" : modulo_navegacao,
    "Modulo de Comunicacao": modulo_comunicacao
}

for nome, status in modulos.items():
    if status == 1:
        print(f"  [OK] {nome:<24}: {status}  (operacional)")
    else:
        print(f"  [FALHA] {nome:<20}: {status}  FALHA!")
        falhas.append(nome)

print("-" * 65)

# ============================================================
# Decisao Final
# ============================================================
print("\n" + "=" * 65)
print("   RESULTADO FINAL")
print("=" * 65)

if len(falhas) == 0:
    print("")
    print("  STATUS --> PRONTO PARA DECOLAR")
    print("  Todos os sistemas estao dentro dos parametros seguros.")
    print("  Autorizacao de decolagem concedida.")
else:
    print("")
    print("  STATUS --> DECOLAGEM ABORTADA")
    print(f"  {len(falhas)} sistema(s) com anomalia detectada:")
    for falha in falhas:
        print(f"    >> {falha}")

print("=" * 65)
````

## 1.4 – Análise Energética

A análise energética calcula a **autonomia inicial** da nave Aurora, considerando a energia disponível antes da decolagem e o consumo estimado durante o lançamento.

````python
# ============================================================
# 1.4 – Analise Energetica
# ============================================================

print("=" * 65)
print("   NAVE AURORA – ANALISE ENERGETICA")
print("=" * 65)

# Parametros energeticos da nave Aurora
capacidade_total_kwh   = 450.0   # kWh – capacidade total dos bancos de bateria
carga_atual_pct        = nivel_energia / 100.0  # fracao decimal
consumo_decolagem_kwh  = 85.0    # kWh – consumo estimado durante o lancamento
taxa_perdas_energeticas = 0.05   # 5% – perdas termicas e eletricas
consumo_medio_missao   = 12.0    # kWh/h – consumo medio em orbita

# --- Calculo 1: Energia disponivel antes da decolagem ---
energia_disponivel = capacidade_total_kwh * carga_atual_pct

# --- Calculo 2: Perdas energeticas estimadas na decolagem ---
perdas_kwh = consumo_decolagem_kwh * taxa_perdas_energeticas

# --- Calculo 3: Energia restante apos a decolagem ---
energia_pos_decolagem = energia_disponivel - consumo_decolagem_kwh - perdas_kwh

# --- Calculo 4: Autonomia estimada em horas ---
autonomia_horas = energia_pos_decolagem / consumo_medio_missao

# Exibicao dos resultados
print(f"\n  Capacidade total da bateria  : {capacidade_total_kwh:.1f} kWh")
print(f"  Carga atual                  : {nivel_energia:.1f}%")
print(f"  Energia disponivel           : {capacidade_total_kwh:.1f} x {carga_atual_pct:.3f} = {energia_disponivel:.2f} kWh")
print("")
print(f"  Consumo estimado (decolagem) : {consumo_decolagem_kwh:.1f} kWh")
print(f"  Perdas energeticas (5%)      : {consumo_decolagem_kwh:.1f} x 0,05 = {perdas_kwh:.2f} kWh")
print(f"  Energia restante apos decolagem: {energia_disponivel:.2f} - {consumo_decolagem_kwh:.1f} - {perdas_kwh:.2f} = {energia_pos_decolagem:.2f} kWh")
print("")
print(f"  Consumo medio em missao      : {consumo_medio_missao:.1f} kWh/h")
print(f"  Autonomia estimada           : {energia_pos_decolagem:.2f} / {consumo_medio_missao:.1f} = {autonomia_horas:.1f} horas")

print("\n" + "-" * 65)
if energia_pos_decolagem > 0 and autonomia_horas >= 20:
    print(f"  ANALISE: Autonomia ADEQUADA para a missao ({autonomia_horas:.1f}h disponíveis).")
elif energia_pos_decolagem > 0:
    print(f"  ANALISE: Autonomia LIMITADA. Apenas {autonomia_horas:.1f}h disponíveis. Monitorar consumo.")
else:
    print("  ANALISE: CRITICO – Energia insuficiente para decolagem segura!")
print("=" * 65)
````

## 1.5 – Análise Assistida por Inteligência Artificial

Os dados de telemetria foram submetidos a um sistema de IA para classificação, detecção de anomalias e sugestão de riscos. A seguir, o relatório gerado:

````python
# ============================================================
# 1.5 – Analise Assistida por IA
# ============================================================

print("=" * 65)
print("   NAVE AURORA – ANALISE ASSISTIDA POR IA")
print("=" * 65)

# Classificacao dos dados pelo sistema de IA
dados_telemetria = {
    "Temperatura Interna"    : {"valor": temperatura_interna,    "min": TEMP_INT_MIN, "max": TEMP_INT_MAX,  "unidade": "C"},
    "Temperatura Externa"    : {"valor": temperatura_externa,    "min": TEMP_EXT_MIN, "max": TEMP_EXT_MAX,  "unidade": "C"},
    "Nivel de Energia"       : {"valor": nivel_energia,          "min": ENERGIA_MIN,  "max": 100.0,         "unidade": "%"},
    "Pressao dos Tanques"    : {"valor": pressao_tanques,        "min": PRESSAO_MIN,  "max": PRESSAO_MAX,   "unidade": "kPa"},
}

print("\n[IA] CLASSIFICACAO DOS DADOS DE TELEMETRIA:")
print("-" * 65)

anomalias = []

for parametro, dados in dados_telemetria.items():
    valor = dados["valor"]
    minv  = dados["min"]
    maxv  = dados["max"]
    amplitude = maxv - minv
    posicao_relativa = (valor - minv) / amplitude * 100  # posicao percentual dentro da faixa

    if posicao_relativa < 0 or posicao_relativa > 100:
        classe = "CRITICO"
        anomalias.append((parametro, valor, "Fora da faixa operacional"))
    elif posicao_relativa < 10 or posicao_relativa > 90:
        classe = "ATENCAO"
        anomalias.append((parametro, valor, "Proximo ao limite da faixa operacional"))
    else:
        classe = "NOMINAL"

    print(f"  {parametro:<25}: {valor:>8} {dados['unidade']:<4} | Posicao na faixa: {posicao_relativa:>5.1f}% | [{classe}]")

# Modulos criticos
print("")
for nome, status in modulos.items():
    classe = "NOMINAL" if status == 1 else "CRITICO"
    if status != 1:
        anomalias.append((nome, status, "Modulo inoperante"))
    print(f"  {nome:<25}: {status}  (1=Operacional)          | [{classe}]")

# Anomalias identificadas
print("\n[IA] ANOMALIAS IDENTIFICADAS:")
print("-" * 65)
if len(anomalias) == 0:
    print("  Nenhuma anomalia critica ou de atencao detectada.")
else:
    for anomalia in anomalias:
        print(f"  >> {anomalia[0]}: valor={anomalia[1]} | Motivo: {anomalia[2]}")

# Sugestoes de risco
print("\n[IA] SUGESTOES DE RISCO:")
print("-" * 65)

# Analise de temperatura interna (prox ao limite superior)
margem_temp_int = TEMP_INT_MAX - temperatura_interna
if margem_temp_int < 5:
    print(f"  [RISCO MEDIO] Temperatura interna ({temperatura_interna}C) a {margem_temp_int:.1f}C do limite superior.")
    print("                Recomendado: verificar sistema de refrigeracao antes da decolagem.")
else:
    print(f"  [BAIXO] Temperatura interna com margem confortavel de {margem_temp_int:.1f}C ate o limite.")

# Analise de energia
print(f"  [BAIXO] Nivel de energia em {nivel_energia}% – acima do minimo de 70%.")
print(f"          Autonomia calculada de {autonomia_horas:.1f}h e adequada para missoes de ate 20h.")

# Analise de pressao dos tanques
margem_pressao_sup = PRESSAO_MAX - pressao_tanques
margem_pressao_inf = pressao_tanques - PRESSAO_MIN
print(f"  [BAIXO] Pressao dos tanques em {pressao_tanques} kPa – centralizada na faixa operacional.")
print(f"          Margem inferior: {margem_pressao_inf:.0f} kPa | Margem superior: {margem_pressao_sup:.0f} kPa.")

print("\n[IA] VEREDICTO GERAL: TODOS OS SISTEMAS CLASSIFICADOS COMO NOMINAIS.")
print("     Missao pode prosseguir com monitoramento continuo da temperatura interna.")
print("=" * 65)
````
